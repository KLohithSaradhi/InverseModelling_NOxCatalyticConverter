```python
import json 
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("./histories/model_lr0.001_nl2_p64_wd0.0_history.json") as f:
    data = json.load(f)
```

```python
plt.plot(np.log10(data["train_loss"]), label="Train loss", alpha=0.7)
plt.plot(np.log10(data["val_loss"]), label = "val loss", alpha=0.7)
plt.legend()
```
