```python
def denorm(norm_params_path, model_output):

    with open(norm_params_path) as f:
        norm_params = json.load(f)

    data = torch.zeros_like(model_output)
    print(data.shape)
    for n in norm_params:
        MAX = norm_params[n]['max']
        MIN = norm_params[n]['min']
        data[:,norm_params[n]['index']] = model_output[:,norm_params[n]['index']] * (MAX - MIN) + MIN

    return data
```
