# Bayesian Inversion with the MLP-PCA Surrogate

This notebook infers the posterior distribution of the six SCR kinetic parameters using the trained MLP-PCA surrogate.

The forward surrogate is

$$
A \longmapsto \tilde{A}
\longmapsto \mathcal{N}_{\theta}(\tilde{A})
= \hat{\tilde{U}}
\longmapsto \hat{U}
\longmapsto \hat{Y}.
$$

Equivalently,

$$
S_{\theta}(A)
=
Y_{\text{mean}}
+
\left[
U_{\text{mean}}
+
U_{\text{std}} \odot
\mathcal{N}_{\theta}
\left(
\frac{A-A_{\text{mean}}}{A_{\text{std}}}
\right)
\right]\Phi.
$$

The Bayesian inverse problem is

$$
p(A \mid y_{\text{obs}})
\propto
p(y_{\text{obs}} \mid A)\,p(A).
$$

For this first clean notebook, we use a synthetic observation from the test set. That lets us check whether the posterior contains the true parameter vector.


## 1. Imports and Paths

Run this notebook from inside `4_BayesInv/`.


```python
from pathlib import Path
import sys
import json

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

# Project layout
ROOT = Path.cwd().parent if Path.cwd().name == "4_BayesInv" else Path.cwd()

DATA_DIR = ROOT / "1_Dataset"
PCA_DIR = ROOT / "2_PCA"
ML_DIR = ROOT / "3_ML"
BAYES_DIR = ROOT / "4_BayesInv"
RESULTS_DIR = BAYES_DIR / "results"
FIGURES_DIR = BAYES_DIR / "figures"

RESULTS_DIR.mkdir(parents=True, exist_ok=True)
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

FULL_DATA_PATH = DATA_DIR / "Full_Dataset.npy"
A_DATA_PATH = DATA_DIR / "lhs_sampling.npy"
T_DATA_PATH = DATA_DIR / "scr_time.npy"

PCA_SPLIT_PATH = PCA_DIR / "PCA_Dataset_r20_split.npz"

MODEL_PREFIX = "mlp_Anorm_to_Unorm_r20"
MODEL_PATH = ML_DIR / f"{MODEL_PREFIX}.pt"
SCALER_PATH = ML_DIR / f"{MODEL_PREFIX}_scalers.npz"
CONFIG_PATH = ML_DIR / f"{MODEL_PREFIX}_config.json"

for p in [FULL_DATA_PATH, A_DATA_PATH, T_DATA_PATH, PCA_SPLIT_PATH, MODEL_PATH, SCALER_PATH, CONFIG_PATH]:
    print(f"{p.name:35s}", "FOUND" if p.exists() else "MISSING", " | ", p)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("\nDevice:", device)
```

## 2. Load Dataset, PCA Split, and MLP Assets

Notation used here:

$$
A = 	ext{kinetic parameter matrix},
\qquad
U = 	ext{PCA coefficient matrix},
\qquad
\Phi = 	ext{PCA basis},
\qquad
Y_{	ext{mean}} = 	ext{mean output signal}.
$$


```python
# Full signal dataset and time grid
Y_full = np.load(FULL_DATA_PATH)
A_all_raw = np.load(A_DATA_PATH)
t = np.load(T_DATA_PATH).squeeze()

# PCA split dataset
split = np.load(PCA_SPLIT_PATH)

A_train = split["A_train"]
A_val = split["A_val"]
A_test = split["A_test"]

U_train = split["U_train"]
U_val = split["U_val"]
U_test = split["U_test"]

idx_train = split["idx_train"]
idx_val = split["idx_val"]
idx_test = split["idx_test"]

Phi = split["Phi"]
Y_mean = split["Y_mean"]
r = int(split["r"])

Y_train = Y_full[idx_train]
Y_val = Y_full[idx_val]
Y_test = Y_full[idx_test]

# Scalers used during MLP training
scalers = np.load(SCALER_PATH)
A_mean = scalers["A_mean"]
A_std = scalers["A_std"]
U_mean = scalers["U_mean"]
U_std = scalers["U_std"]

with open(CONFIG_PATH, "r") as f:
    config = json.load(f)

print("Y_full:", Y_full.shape)
print("A_all_raw:", A_all_raw.shape)
print("t:", t.shape)
print("Phi:", Phi.shape)
print("Y_mean:", Y_mean.shape)
print("r:", r)
print()
print("A_train/A_val/A_test:", A_train.shape, A_val.shape, A_test.shape)
print("U_train/U_val/U_test:", U_train.shape, U_val.shape, U_test.shape)
print()
print("Config:")
print(json.dumps(config, indent=2))
```

## 3. Rebuild the Trained MLP Surrogate

The MLP maps normalized parameters to normalized PCA coefficients:

$$
\mathcal{N}_{	heta}: 	ilde{A} \mapsto 	ilde{U}.
$$


```python
def get_activation(name):
    name = str(name).lower()
    if name == "tanh":
        return nn.Tanh()
    if name == "relu":
        return nn.ReLU()
    if name == "gelu":
        return nn.GELU()
    if name == "silu":
        return nn.SiLU()
    raise ValueError(f"Unknown activation: {name}")


class MLP(nn.Module):
    def __init__(self, in_dim, out_dim, hidden_dim=128, num_layers=4, activation="tanh"):
        super().__init__()

        layers = []
        current_dim = in_dim

        for _ in range(num_layers):
            layers.append(nn.Linear(current_dim, hidden_dim))
            layers.append(get_activation(activation))
            current_dim = hidden_dim

        layers.append(nn.Linear(current_dim, out_dim))
        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)


def get_config_value(config, key, default):
    if key in config:
        return config[key]
    if "model" in config and key in config["model"]:
        return config["model"][key]
    return default


in_dim = int(get_config_value(config, "in_dim", A_train.shape[1]))
out_dim = int(get_config_value(config, "out_dim", U_train.shape[1]))
hidden_dim = int(get_config_value(config, "hidden_dim", 128))
num_layers = int(get_config_value(config, "num_layers", 4))
activation = get_config_value(config, "activation", "tanh")

mlp = MLP(
    in_dim=in_dim,
    out_dim=out_dim,
    hidden_dim=hidden_dim,
    num_layers=num_layers,
    activation=activation,
).to(device)

try:
    checkpoint = torch.load(MODEL_PATH, map_location=device, weights_only=True)
except TypeError:
    checkpoint = torch.load(MODEL_PATH, map_location=device)

if isinstance(checkpoint, dict) and "model_state_dict" in checkpoint:
    state_dict = checkpoint["model_state_dict"]
elif isinstance(checkpoint, dict) and "state_dict" in checkpoint:
    state_dict = checkpoint["state_dict"]
else:
    state_dict = checkpoint

mlp.load_state_dict(state_dict)
mlp.eval()

print(mlp)
print("\nLoaded model from:", MODEL_PATH)
```

## 4. Define the Forward Surrogate

For any physical parameter vector $A$, this evaluates

$$
\hat{Y} = S_{	heta}(A).
$$


```python
def ensure_2d(X):
    X = np.asarray(X, dtype=float)
    if X.ndim == 1:
        X = X[None, :]
    return X


def predict_U(A):
    """Predict physical PCA coefficients U from physical kinetic parameters A."""
    A = ensure_2d(A)
    A_norm = (A - A_mean) / A_std

    with torch.no_grad():
        A_tensor = torch.tensor(A_norm, dtype=torch.float32, device=device)
        U_norm = mlp(A_tensor).cpu().numpy()

    U_pred = U_mean + U_std * U_norm
    return U_pred


def reconstruct_Y_from_U(U):
    """Reconstruct full output signal from physical PCA coefficients."""
    U = ensure_2d(U)
    return U @ Phi + Y_mean


def predict_Y(A):
    """Full MLP-PCA surrogate: A -> Y."""
    U_pred = predict_U(A)
    return reconstruct_Y_from_U(U_pred)


# Quick test on a few held-out samples
Y_hat_test_small = predict_Y(A_test[:5])
print("Y_hat_test_small:", Y_hat_test_small.shape)
```

## 5. Select a Synthetic Observation

We choose one test sample as the observed signal:

$$
y_{	ext{obs}}(t) = y_{	ext{test}}(t) + \epsilon(t).
$$

Set `OBS_NOISE_STD = 0.0` for a noise-free synthetic inverse problem.


```python
OBS_TEST_ID = 0
OBS_NOISE_STD = 0.0
OBS_SEED = 123

rng = np.random.default_rng(OBS_SEED)

A_true = A_test[OBS_TEST_ID].copy()
y_clean = Y_test[OBS_TEST_ID].copy()
y_obs = y_clean + OBS_NOISE_STD * rng.normal(size=y_clean.shape)

y_sur_true = predict_Y(A_true).squeeze()

print("OBS_TEST_ID:", OBS_TEST_ID)
print("A_true:")
for j, value in enumerate(A_true):
    print(f"A[{j}] = {value:.8e}")

print("\ny_obs:", y_obs.shape)
print("y_obs min/max:", y_obs.min(), y_obs.max())
```

```python
plt.figure(figsize=(10, 4))
plt.plot(t, y_obs, linewidth=1.8, label="observed signal")
plt.plot(t, y_sur_true, linestyle="--", linewidth=1.4, label="surrogate at true A")
plt.xlabel("time")
plt.ylabel("SCR output")
plt.title("Synthetic observation")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 6. Parameter Bounds and Normalized Inversion Variables

MCMC is run in normalized coordinates

$$
z \in [0,1]^6.
$$

The physical parameters are recovered by

$$
A = A_{\min} + z \odot (A_{\max}-A_{\min}).
$$


```python
# Use the sampled dataset bounds as the prior box.
A_all = np.vstack([A_train, A_val, A_test])

A_lower = A_all.min(axis=0)
A_upper = A_all.max(axis=0)
A_range = A_upper - A_lower

if np.any(A_range <= 0):
    raise ValueError("At least one parameter has zero or negative prior range.")

A_NAMES = [
    "k_ads",
    "k_des",
    "k_std",
    "k_fst",
    "k_slw",
    "k_oxi",
]

bounds_df = pd.DataFrame({
    "parameter": A_NAMES,
    "lower": A_lower,
    "upper": A_upper,
    "true_A": A_true,
})

bounds_df
```

```python
def z_to_A(z):
    z = np.asarray(z, dtype=float)
    return A_lower + z * A_range


def A_to_z(A):
    A = np.asarray(A, dtype=float)
    return (A - A_lower) / A_range


z_true = A_to_z(A_true)

print("z_true:")
print(z_true)
```

## 7. Define the Log Posterior

We use a uniform prior over the normalized box:

$$
p(z)=
egin{cases}
1, & 0 \le z_j \le 1 \quad orall j, \
0, & 	ext{otherwise}.
\end{cases}
$$

For the likelihood, assume independent Gaussian residuals:

$$
y_{	ext{obs}}(t_i)
=
S_{	heta}(A(z))(t_i)+\epsilon_i,
\qquad
\epsilon_i \sim \mathcal{N}(0,\sigma^2).
$$

Therefore,

$$
\log p(y_{	ext{obs}}\mid z)
=
-rac{1}{2}\sum_i
\left(
rac{y_{	ext{obs}}(t_i)-S_{	heta}(A(z))(t_i)}{\sigma}
ight)^2
-N_t\log\sigma
+C.
$$


```python
# Likelihood noise scale.
# For the first synthetic test, a simple fixed scale is enough.
LIKELIHOOD_REL_SIGMA = 0.01
LIKELIHOOD_SIGMA_FLOOR = 1e-8

sigma = max(LIKELIHOOD_REL_SIGMA * np.std(y_obs), LIKELIHOOD_SIGMA_FLOOR)

print("sigma:", sigma)
```

```python
def log_prior_z(z):
    z = np.asarray(z, dtype=float)
    if np.all((z >= 0.0) & (z <= 1.0)):
        return 0.0
    return -np.inf


def log_likelihood_z(z):
    A = z_to_A(z)
    y_pred = predict_Y(A).squeeze()
    residual = y_obs - y_pred
    return -0.5 * np.sum((residual / sigma)**2) - len(y_obs) * np.log(sigma)


def log_posterior_z(z):
    lp = log_prior_z(z)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood_z(z)


print("log posterior at true z:  ", log_posterior_z(z_true))
print("log posterior at center z:", log_posterior_z(np.full(6, 0.5)))
```

## 8. Random-Walk Metropolis Sampler

Proposal:

$$
z^{\star} = z^{(m)} + \eta,
\qquad
\eta \sim \mathcal{N}(0, s^2 I).
$$

Acceptance probability:

$$
lpha
=
\min\left(1,
\exp\left[\log p(z^{\star}\mid y_{	ext{obs}})
-
\log p(z^{(m)}\mid y_{	ext{obs}})
ight]
ight).
$$


```python
def run_random_walk_mcmc(log_post_fn, z_init, n_steps, proposal_scale, seed):
    rng = np.random.default_rng(seed)

    z_current = np.asarray(z_init, dtype=float).copy()
    logp_current = log_post_fn(z_current)

    n_dim = len(z_current)
    chain = np.zeros((n_steps, n_dim))
    logp = np.zeros(n_steps)
    accepted = np.zeros(n_steps, dtype=bool)

    for i in range(n_steps):
        z_prop = z_current + proposal_scale * rng.normal(size=n_dim)
        logp_prop = log_post_fn(z_prop)

        log_alpha = logp_prop - logp_current

        if np.log(rng.uniform()) < log_alpha:
            z_current = z_prop
            logp_current = logp_prop
            accepted[i] = True

        chain[i] = z_current
        logp[i] = logp_current

    return chain, logp, accepted.mean()
```

## 9. Pilot MCMC

The pilot chain is only for checking proposal scale and basic sampler behavior.


```python
PILOT_STEPS = 3000
PILOT_PROPOSAL_SCALE = 0.02

# For synthetic debugging, starting near the truth is useful.
# Change this to np.full(6, 0.5) or a random point for a harder inversion test.
z_init_pilot = np.clip(z_true + 0.05 * np.random.default_rng(1).normal(size=6), 0.0, 1.0)

pilot_chain, pilot_logp, pilot_accept = run_random_walk_mcmc(
    log_post_fn=log_posterior_z,
    z_init=z_init_pilot,
    n_steps=PILOT_STEPS,
    proposal_scale=PILOT_PROPOSAL_SCALE,
    seed=10,
)

best_pilot_idx = int(np.argmax(pilot_logp))
z_best_pilot = pilot_chain[best_pilot_idx]

print("Pilot acceptance rate:", pilot_accept)
print("Best pilot log posterior:", pilot_logp.max())
print("Best pilot z:", z_best_pilot)
```

```python
plt.figure(figsize=(9, 3))
plt.plot(pilot_logp, linewidth=1.0)
plt.xlabel("MCMC step")
plt.ylabel("log posterior")
plt.title("Pilot log posterior trace")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 10. Production MCMC

After burn-in, the remaining samples approximate draws from

$$
p(A\mid y_{	ext{obs}}).
$$


```python
N_STEPS = 20000
PROPOSAL_SCALE = PILOT_PROPOSAL_SCALE
BURN_IN = 3000

prod_chain, prod_logp, prod_accept = run_random_walk_mcmc(
    log_post_fn=log_posterior_z,
    z_init=z_best_pilot,
    n_steps=N_STEPS,
    proposal_scale=PROPOSAL_SCALE,
    seed=20,
)

z_samples = prod_chain[BURN_IN:]
logp_samples = prod_logp[BURN_IN:]
A_samples = np.array([z_to_A(z) for z in z_samples])

print("Production acceptance rate:", prod_accept)
print("Total samples:", len(prod_chain))
print("Burn-in:", BURN_IN)
print("Kept samples:", len(z_samples))
print("Best production log posterior:", prod_logp.max())
```

```python
plt.figure(figsize=(9, 3))
plt.plot(prod_logp, linewidth=0.8)
plt.axvline(BURN_IN, linestyle="--", linewidth=1.2, label="burn-in")
plt.xlabel("MCMC step")
plt.ylabel("log posterior")
plt.title("Production log posterior trace")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(6, 1, figsize=(10, 9), sharex=True)

for j, ax in enumerate(axes):
    ax.plot(prod_chain[:, j], linewidth=0.7)
    ax.axhline(z_true[j], linestyle="--", linewidth=1.0)
    ax.set_ylabel(A_NAMES[j])
    ax.set_ylim(-0.05, 1.05)
    ax.grid(True)

axes[-1].set_xlabel("MCMC step")
fig.suptitle("Production traces in normalized z-space", y=1.01)
plt.tight_layout()
plt.show()
```

## 11. Posterior Parameter Summary

For each parameter, we report posterior mean, median, and the 95% credible interval.


```python
def summarize_samples(samples, names, true_values=None):
    rows = []
    for j, name in enumerate(names):
        row = {
            "parameter": name,
            "mean": np.mean(samples[:, j]),
            "median": np.median(samples[:, j]),
            "q025": np.quantile(samples[:, j], 0.025),
            "q975": np.quantile(samples[:, j], 0.975),
        }
        if true_values is not None:
            row["true"] = true_values[j]
        rows.append(row)
    return pd.DataFrame(rows)


z_summary = summarize_samples(z_samples, A_NAMES, true_values=z_true)
A_summary = summarize_samples(A_samples, A_NAMES, true_values=A_true)

A_summary
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
axes = axes.ravel()

for j, ax in enumerate(axes):
    ax.hist(z_samples[:, j], bins=40, density=True, alpha=0.8)
    ax.axvline(z_true[j], linestyle="--", linewidth=1.5, label="true")
    ax.axvline(z_summary.loc[j, "median"], linestyle=":", linewidth=1.5, label="median")
    ax.set_title(A_NAMES[j])
    ax.set_xlabel("z")
    ax.set_ylabel("density")
    ax.set_xlim(0.0, 1.0)
    ax.grid(True)

axes[0].legend()
plt.tight_layout()
plt.show()
```

## 12. Posterior Predictive Check

Push posterior samples through the surrogate:

$$
\hat{y}^{(s)}(t) = S_{	heta}(A^{(s)})(t).
$$

Then compute the posterior predictive mean and 95% band:

$$
ar{y}(t) = rac{1}{N_s}\sum_{s=1}^{N_s}\hat{y}^{(s)}(t).
$$


```python
N_PRED_SAMPLES = min(2000, len(A_samples))
rng = np.random.default_rng(30)
pred_ids = rng.choice(len(A_samples), size=N_PRED_SAMPLES, replace=False)

Y_post = predict_Y(A_samples[pred_ids])

y_post_mean = Y_post.mean(axis=0)
y_post_q025 = np.quantile(Y_post, 0.025, axis=0)
y_post_q975 = np.quantile(Y_post, 0.975, axis=0)

res = y_obs - y_post_mean
rmse = np.sqrt(np.mean(res**2))
rel_rmse = np.linalg.norm(res) / np.linalg.norm(y_obs)
mae = np.mean(np.abs(res))
max_abs = np.max(np.abs(res))

print("Posterior predictive errors")
print("RMSE:         ", rmse)
print("Relative RMSE:", rel_rmse)
print("MAE:          ", mae)
print("Max abs error:", max_abs)
```

```python
plt.figure(figsize=(10, 4))
plt.fill_between(t, y_post_q025, y_post_q975, alpha=0.25, label="95% posterior predictive band")
plt.plot(t, y_obs, linewidth=1.8, label="observed signal")
plt.plot(t, y_post_mean, linewidth=1.8, label="posterior mean prediction")
plt.xlabel("time")
plt.ylabel("SCR output")
plt.title(f"Posterior predictive check: RMSE = {rmse:.4g}, rel. RMSE = {rel_rmse:.4g}")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 13. Save Posterior Results


```python
posterior_npz_path = RESULTS_DIR / "posterior_mcmc_synthetic_test0.npz"
summary_csv_path = RESULTS_DIR / "posterior_summary_synthetic_test0.csv"

np.savez(
    posterior_npz_path,
    z_samples=z_samples,
    A_samples=A_samples,
    logp_samples=logp_samples,
    z_true=z_true,
    A_true=A_true,
    y_obs=y_obs,
    y_post_mean=y_post_mean,
    y_post_q025=y_post_q025,
    y_post_q975=y_post_q975,
    t=t,
    sigma=sigma,
    obs_test_id=OBS_TEST_ID,
)

A_summary.to_csv(summary_csv_path, index=False)

print("Saved:", posterior_npz_path)
print("Saved:", summary_csv_path)
```

## Notes

- `RMSE` is the average absolute error scale in the original output units.
- `Relative RMSE` is $\|y-\hat{y}\|_2 / \|y\|_2$. A value of `0.01` means the prediction error norm is about 1% of the signal norm. It is not exactly the pointwise ratio $(y(t)-\hat{y}(t))/\hat{y}(t)$ at every time.
- `MAE` is the average absolute pointwise error.
