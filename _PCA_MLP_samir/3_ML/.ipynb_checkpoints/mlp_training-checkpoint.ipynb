# MLP Surrogate Training for PCA Coefficients

This notebook trains a neural-network surrogate for the LSOPT inverse-modeling workflow.

From the PCA compression notebook, each SCR model output signal is represented approximately as

$$
y^{(m)} \approx \bar{y} + \Phi_r a^{(m)}
$$

where:

- $k^{(m)} \in \mathbb{R}^6$ is the sampled kinetic-parameter vector,
- $y^{(m)} \in \mathbb{R}^{N_t}$ is the corresponding SCR output time series,
- $\bar{y}$ is the mean signal,
- $\Phi_r$ contains the first $r$ PCA modes,
- $a^{(m)} \in \mathbb{R}^r$ is the vector of PCA coefficients.

The PCA compression step already produced the reduced coefficients

$$
a^{(m)}, \qquad m = 1,\dots,N_s.
$$

The goal of this notebook is to train an MLP surrogate

$$
\mathcal{N}_\theta : k^{(m)} \longmapsto a^{(m)}
$$

so that for a new parameter vector $k^\star$, we can predict

$$
\hat{a}^\star = \mathcal{N}_\theta(k^\star)
$$

and reconstruct the SCR output signal as

$$
\hat{y}^\star \approx \bar{y} + \Phi_r \hat{a}^\star.
$$

In the saved PCA artifact:

- `A` stores the coefficient vectors $a^{(m)}$ as rows, with shape `(1000, 20)`,
- `pca_components` stores the PCA basis $\Phi_r$ in transposed row format, with shape `(20, 11993)`,
- `Y_mean` stores $\bar{y}$, with shape `(11993,)`,
- `t` stores the output time grid, with shape `(11993,)`.

Everything stays local inside the `_LSOPT` folder for now.

## 3. Normalize Inputs and PCA Coefficients

This step normalizes the input parameters $k^{(m)}$ and PCA coefficients $a^{(m)}$ using statistics computed from the training set only.

This avoids leaking validation or test information into the training process.

The normalized quantities are

$$
\tilde{k} = \frac{k - \mu_k}{\sigma_k}
$$

and

$$
\tilde{a} = \frac{a - \mu_a}{\sigma_a}.
$$

The MLP is trained on the normalized map

$$
\tilde{k}^{(m)} \longmapsto \tilde{a}^{(m)}.
$$

````python
# Compute normalization statistics from training data only
K_mean = K_train.mean(axis=0)
K_std = K_train.std(axis=0) + 1e-12

A_mean = A_train.mean(axis=0)
A_std = A_train.std(axis=0) + 1e-12

def normalize_K(K_raw):
    return (K_raw - K_mean) / K_std

def normalize_A(A_raw):
    return (A_raw - A_mean) / A_std

def denormalize_A(A_norm):
    return A_norm * A_std + A_mean

# Apply normalization
K_train_n = normalize_K(K_train)
K_val_n = normalize_K(K_val)
K_test_n = normalize_K(K_test)

A_train_n = normalize_A(A_train)
A_val_n = normalize_A(A_val)
A_test_n = normalize_A(A_test)

print("K_train_n mean:", K_train_n.mean(axis=0))
print("K_train_n std: ", K_train_n.std(axis=0))
print()
print("A_train_n mean first 5:", A_train_n.mean(axis=0)[:5])
print("A_train_n std first 5: ", A_train_n.std(axis=0)[:5])
````

## 4. Build PyTorch Datasets and DataLoaders

This step converts the normalized NumPy arrays into PyTorch tensors.

The training set is shuffled during each epoch. The validation and test sets are not shuffled.

````python
batch_size = 64

train_ds = TensorDataset(
    torch.tensor(K_train_n, dtype=torch.float32),
    torch.tensor(A_train_n, dtype=torch.float32),
)

val_ds = TensorDataset(
    torch.tensor(K_val_n, dtype=torch.float32),
    torch.tensor(A_val_n, dtype=torch.float32),
)

test_ds = TensorDataset(
    torch.tensor(K_test_n, dtype=torch.float32),
    torch.tensor(A_test_n, dtype=torch.float32),
)

train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=batch_size, shuffle=False)

xb, yb = next(iter(train_loader))

print("batch_size:", batch_size)
print("Input batch shape:", xb.shape)
print("Output batch shape:", yb.shape)
````

## 5. Define the MLP Surrogate

This step defines the neural-network surrogate.

The model takes normalized kinetic parameters $\tilde{k}^{(m)} \in \mathbb{R}^6$ as input and predicts normalized PCA coefficients $\tilde{a}^{(m)} \in \mathbb{R}^{20}$ as output:

$$
\mathcal{N}_{\theta}: \tilde{k}^{(m)} \longmapsto \tilde{a}^{(m)}.
$$

We use a fully connected multilayer perceptron with `Tanh` activations.

````python
class MLP(nn.Module):
    def __init__(self, in_dim, out_dim, hidden_dim=128, num_layers=4):
        super().__init__()

        layers = []
        current_dim = in_dim

        for _ in range(num_layers):
            layers.append(nn.Linear(current_dim, hidden_dim))
            layers.append(nn.Tanh())
            current_dim = hidden_dim

        layers.append(nn.Linear(current_dim, out_dim))

        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)


model = MLP(
    in_dim=k_dim,
    out_dim=r,
    hidden_dim=128,
    num_layers=4,
).to(device)

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=1e-3,
    weight_decay=1e-5,
)

loss_fn = nn.MSELoss()

n_params = sum(p.numel() for p in model.parameters())

print(model)
print()
print("Number of trainable parameters:", n_params)
````

## 6. Train the MLP with Early Stopping

This step trains the MLP using mean-squared error loss on the normalized PCA coefficients.

At each epoch, the model is trained on the training set and evaluated on the validation set. The best model is selected using the validation loss.

Early stopping is used to stop training once the validation loss stops improving.

````python
def evaluate(model, loader):
    model.eval()
    total_loss = 0.0
    total_count = 0

    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            yb = yb.to(device)

            pred = model(xb)
            loss = loss_fn(pred, yb)

            total_loss += loss.item() * xb.shape[0]
            total_count += xb.shape[0]

    return total_loss / total_count
````

````python
max_epochs = 3000
patience = 300

best_val_loss = np.inf
best_epoch = -1
wait = 0

history = {
    "train_loss": [],
    "val_loss": [],
}

for epoch in range(1, max_epochs + 1):
    model.train()

    train_loss_sum = 0.0
    train_count = 0

    for xb, yb in train_loader:
        xb = xb.to(device)
        yb = yb.to(device)

        optimizer.zero_grad()

        pred = model(xb)
        loss = loss_fn(pred, yb)

        loss.backward()
        optimizer.step()

        train_loss_sum += loss.item() * xb.shape[0]
        train_count += xb.shape[0]

    train_loss = train_loss_sum / train_count
    val_loss = evaluate(model, val_loader)

    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_epoch = epoch
        wait = 0

        torch.save(
            {
                "model_state_dict": model.state_dict(),
                "k_dim": k_dim,
                "r": r,
                "hidden_dim": 128,
                "num_layers": 4,
                "best_val_loss": best_val_loss,
                "best_epoch": best_epoch,
                "model_type": "MLP",
                "input_type": "normalized_k",
                "output_type": "normalized_pca_coefficients",
            },
            MODEL_PATH,
        )
    else:
        wait += 1

    if epoch == 1 or epoch % 100 == 0:
        print(
            f"epoch {epoch:5d} | "
            f"train {train_loss:.4e} | "
            f"val {val_loss:.4e} | "
            f"best {best_val_loss:.4e}"
        )

    if wait >= patience:
        print()
        print(f"Early stopping at epoch {epoch}")
        print(f"Best epoch: {best_epoch}")
        print(f"Best val loss: {best_val_loss:.4e}")
        break
````

## 7. Save Training Metadata

This step saves the normalization scalers, training history, and model configuration.

The model checkpoint already stores the neural-network weights. The scaler file is needed later because the deployed workflow must apply the same normalization:

$$
k \rightarrow \tilde{k}
$$

before prediction, and then convert

$$
\tilde{a} \rightarrow a
$$

after prediction.

````python
# Save normalization scalers
np.savez(
    SCALER_PATH,
    K_mean=K_mean,
    K_std=K_std,
    A_mean=A_mean,
    A_std=A_std,
)

# Save training history
with open(HISTORY_PATH, "w") as f:
    json.dump(history, f, indent=2)

# Save configuration
config = {
    "model_name": "mlp_kn_to_pcan_r20",
    "model_type": "MLP",
    "input": "normalized kinetic parameters",
    "output": "normalized PCA coefficients",
    "k_dim": int(k_dim),
    "r": int(r),
    "hidden_dim": 128,
    "num_layers": 4,
    "activation": "Tanh",
    "batch_size": int(batch_size),
    "max_epochs": int(max_epochs),
    "patience": int(patience),
    "best_epoch": int(best_epoch),
    "best_val_loss": float(best_val_loss),
    "optimizer": "AdamW",
    "lr": 1e-3,
    "weight_decay": 1e-5,
    "model_path": str(MODEL_PATH),
    "scaler_path": str(SCALER_PATH),
    "pca_path": str(PCA_PATH),
}

with open(CONFIG_PATH, "w") as f:
    json.dump(config, f, indent=2)

print("Saved model:  ", MODEL_PATH)
print("Saved scalers:", SCALER_PATH)
print("Saved history:", HISTORY_PATH)
print("Saved config: ", CONFIG_PATH)
````

````python
with torch.no_grad():
    K_test_t = torch.tensor(K_test_n, dtype=torch.float32).to(device)
    A_test_pred_n = model(K_test_t).cpu().numpy()

# Convert normalized PCA coefficients back to original PCA coefficient scale
A_test_pred = denormalize_A(A_test_pred_n)

rmse_A = np.sqrt(np.mean((A_test_pred - A_test)**2))
rel_rmse_A = np.linalg.norm(A_test_pred - A_test) / np.linalg.norm(A_test)

print("Test PCA coefficient RMSE:        ", rmse_A)
print("Test PCA coefficient relative RMSE:", rel_rmse_A)
````

## 10. Reconstruct Test Output Signals

This step maps the predicted PCA coefficients back to the original SCR output-signal space.

Using the PCA representation,

$$
\hat{y}^{(m)} \approx \bar{y} + \Phi_r \hat{a}^{(m)},
$$

and because `pca_components` is stored row-wise with shape `(r, N_t)`, reconstruction is implemented as:

```python
Y_pred = A_pred @ pca_components + Y_mean

````python
# Reconstruct test signals from true and predicted PCA coefficients
Y_test_pca_true = A_test @ pca_components + Y_mean
Y_test_mlp_pred = A_test_pred @ pca_components + Y_mean

rel_rmse_Y = np.linalg.norm(Y_test_mlp_pred - Y_test_pca_true) / np.linalg.norm(Y_test_pca_true)
rmse_Y = np.sqrt(np.mean((Y_test_mlp_pred - Y_test_pca_true)**2))

print("Test reconstructed signal RMSE:         ", rmse_Y)
print("Test reconstructed signal relative RMSE:", rel_rmse_Y)
print()
print("Y_test_pca_true shape:", Y_test_pca_true.shape)
print("Y_test_mlp_pred shape:", Y_test_mlp_pred.shape)
````

````python
# Plot several test reconstructions: one sample per row
rng = np.random.default_rng(0)
n_show = 6
show_ids = rng.choice(A_test.shape[0], size=n_show, replace=False)

fig, axes = plt.subplots(n_show, 1, figsize=(10, 3.2 * n_show), sharex=True)

# If n_show = 1, make axes iterable
if n_show == 1:
    axes = [axes]

for ax, i in zip(axes, show_ids):
    ax.plot(t, Y_test_pca_true[i], label="PCA target", linewidth=2)
    ax.plot(t, Y_test_mlp_pred[i], "--", label="MLP prediction", linewidth=2)

    ax.set_title(f"Test sample {i}")
    ax.set_ylabel("output")
    ax.grid(True)

axes[0].legend()
axes[-1].set_xlabel("time")

plt.tight_layout()
plt.show()
````

````python
# Signal-space error between PCA target and MLP reconstructed signal
err = Y_test_mlp_pred - Y_test_pca_true

rel_rmse_Y = np.linalg.norm(err) / np.linalg.norm(Y_test_pca_true)
rmse_Y = np.sqrt(np.mean(err**2))

print("Signal-space RMSE:", rmse_Y)
print("Signal-space relative RMSE:", rel_rmse_Y)
````

````python
per_sample_rel_err = np.linalg.norm(err, axis=1) / np.linalg.norm(Y_test_pca_true, axis=1)

print("Per-sample relative error:")
print("min:   ", per_sample_rel_err.min())
print("mean:  ", per_sample_rel_err.mean())
print("median:", np.median(per_sample_rel_err))
print("max:   ", per_sample_rel_err.max())
````

### Interpretation

The MLP has a larger relative error in PCA coefficient space than in reconstructed signal space.

This is expected because coefficient-space error weights all retained PCA coefficients directly, while signal-space reconstruction is dominated by the mean signal and the leading PCA modes. Errors in lower-energy PCA coefficients may have little visible effect on the reconstructed output signal.

For this workflow, the signal-space error is the more important metric because the final surrogate output is reconstructed as

$$
\hat{y}(t) \approx \bar{y} + \Phi_r \hat{a}.
$$

The current baseline MLP gives a test signal-space relative RMSE of about

1.16\%.
